```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def f(x,y):
    return x**2 + y**2 / 4
```

```python
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
xmesh, ymesh = np.meshgrid(x,y)
# xmesh : Matirix of x coordinate for each mesh point
# ymesh : Matirix of y coordinate for each mesh point

z = f(xmesh.ravel(), ymesh.ravel()).reshape(xmesh.shape)
# before broadcast, converse to 1D array

print("shape of xmesh")
print(xmesh.shape)
```

```python
plt.contour(x,y,z, colors="k", levels=[1,2,3,4,5])
```

# Add grayscale color

```python
colors = ["0.1", "0.2" ,"0.3", "0.4"]
plt.contourf(x,y,z, colors=colors, levels=[1,2,3,4,5])
```
