# Polynomial Regression

```python
from scipy import linalg
import numpy as np
```

```python
class LinearRegression:
    def __init__(self):
        self.w_ = None
        
    def fit(self, x, t):
        Xtil = np.c_[np.ones(x.shape[0]), x] # concatecate offset terms
        A = np.dot(Xtil.T, Xtil)
        b = np.dot(Xtil.T, t)
        self.w_ = linalg.solve(A, b)
        
    def predict(self, X):
        if X.ndim  == 1:
            X = X.reshape(-1, 1)
        Xtil = np.c_[np.ones(X.shape[0]), X]
        return np.dot(Xtil, self.w_)
```

```python
class PolynomialRegression:
    def __init__(self, degree):
        self.degree = degree
        
    def fit(self, x, y):
        x_pow = []
        xx = x.reshape(len(x), 1) # to n x 1 column veector
        for i in range(1, self.degree + 1):
            x_pow.append(xx**i)
            
        mat = np.concatenate(x_pow, axis=1)
        linreg = LinearRegression()
        linreg.fit(mat, y)
        self.w_ = linreg.w_
        
    def predict(self, x):
        r = 0
        #for i in range(self.degree + 1):
        #    r += x**i * self.w[i]
        for i in range(self.degree + 1):
            r += x**i * self.w_[i]
        
        return r
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

def f(x):
    return 1 + 2 * x

# generate data including Noise
x = np.random.random(10) * 10
y = f(x) + np.random.randn(10)

# make model
model = PolynomialRegression(10) # 10:degree
model.fit(x, y)

# plot
plt.ylim([y.min() - 1, y.max() + 1])
plt.scatter(x, y, color='k')

xx = np.linspace(x.min(), x.max(), 300)
yy = np.array([model.predict(u) for u in xx])
plt.plot(xx, yy, color='k')

# Linear Regression
model = LinearRegression()
model.fit(x, y)
b, a = model.w_
x1 = x.min() - 1
x2 = x.max() + 1
plt.plot([x1, x2], [f(x1), f(x2)], color='k', linestyle='dashed')
```

# Bias and Variance

```python
import warnings
```

```python
# True func.
def f(x):
    return 1 / (1 + x)

# gererate sample data using true function
def sample(n):
    x = np.random.random(n) * 5
    y = f(x)
    return x, y

xx = np.arange(0, 5, 0.01)
np.random.seed(0)
y_poly_sum = np.zeros(len(xx))
y_poly_sum_sq = np.zeros(len(xx))
y_lin_sum = np.zeros(len(xx))
y_lin_sum_sq = np.zeros(len(xx))
y_true = f(xx)

n = 100000
warnings.filterwarnings("ignore")

for _ in range(n):
    x, y = sample(5)

    poly = PolynomialRegression(4)
    poly.fit(x, y)
    y_poly = poly.predict(xx)
    y_poly_sum += y_poly
    y_poly_sum_sq += (y_poly - y_true)**2 # sum of square error
    
    lin = LinearRegression()
    lin.fit(x, y)
    y_lin = lin.predict(xx)
    y_lin_sum += y_lin
    y_lin_sum_sq += (y_lin - y_true)**2 # sum of square error
    
plt.plot(xx, f(xx), label='truth', color='k', linestyle="solid")
plt.plot(xx, y_poly_sum/n, label='polynomial', color='k', linestyle="dotted")
plt.plot(xx, y_lin_sum/n, label='linear', color='k', linestyle="dashed")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.set_title("Linear reg.")
ax2.set_title("Polynomial reg")
ax1.set_ylim(0,1)
ax2.set_ylim(0,1)
```
